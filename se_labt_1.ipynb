```python
import matplotlib.pyplot as plt
import numpy as np
# hard code variables
def quadratic_model(time):
  a=3
  b=9
  c=27
  temperature = a*time*time + b*time + c
  return temperature

time_values=np.linspace(0,10,100)
temperature_hardcoded = quadratic_model(time_values)

plt.plot(time_values,temperature_hardcoded, label='temperature with hard coded coefficients')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
plt.title('weather modeling with quadratic equation(Hard-coded coefficient)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time**2) + b * time + c
    return temperature

def main():
    coefficient_sets = [
        {'a': 0.4, 'b': -5, 'c': 40},
        {'a': 0.2, 'b': -2, 'c': 50},
        {'a': 0.15, 'b': -0.5, 'c': 25}
    ]

    time_values = np.linspace(0, 10, 100)

    for idx, coefficients in enumerate(coefficient_sets):
        temperature = quadratic_model(time_values, coefficients['a'], coefficients['b'], coefficients['c'])
        label = f'Set {idx + 1} coefficients'
        plt.plot(time_values, temperature, label=label)

        annotation_text = f'Set {idx + 1}: a={coefficients["a"]}, b={coefficients["b"]}, c={coefficients["c"]}'
        plt.annotate(annotation_text, xy=(time_values[-1], temperature[-1]), textcoords="offset points", xytext=(5,0), ha='left')

    plt.xlabel('time')
    plt.ylabel('temperature')
    plt.legend()
    plt.title('Weather Modeling with Quadratic Equation (Multiple Sets of Coefficients)')

    plt.show()

if __name__ == "__main__":
    main()
```
